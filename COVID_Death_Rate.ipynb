```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AnubhaT-code/covid-19-data/master/public/data/latest/owid-covid-latest.csv')
```

```python
myvar = pd.Series(df['iso_code'])
```

```python
deathRate = []

f = 0
posmax = -1
max = -1 
for x in myvar:
    if df['total_cases'].values[f]==0:
        continue
    calc = (df['total_deaths'].values[f]*100.0)/df['total_cases'].values[f]
    if calc > max:
        max = calc
        posmax = f
    deathRate.append(calc)
    f = f+1
```

```python
print(deathRate)
```
